## Coding Gradient Descent

```python
def step_gradient(data,learning_rate,m,c):
    m_slope=0
    c_slope=0
    M=len(data)
    for i in range(M):
        x=data[i,0]
        y=data[i,1]
        m_slope+=(-2/M)*(y-m*x-c)*x
        c_slope+=(-2/M)*(y-m*x-c)
    new_m=m-learning_rate*m_slope
    new_c=c-learning_rate*c_slope
    return new_m,new_c
```

```python
def cost(data,m,c):
    total_cost=0
    M=len(data)
    for i in range(M):
        x=data[i,0]
        y=data[i,1]
        total_cost+=(1/M)*((y-m*x-c)**2)
    return total_cost
```

```python
def gd(data,learning_rate,num_iterations):
    m=0
    c=0
    for i in range(num_iterations):
        m,c=step_gradient(data,learning_rate,m,c)
        print(i,"Cost : ",cost(data,m,c))
    return m,c
```

```python
def run():
    data=np.loadtxt(r"C:\Users\DEEPAK\My ML Codes\Gradient Descent\data.csv",delimiter=",")
    learning_rate=0.0001
    num_iterations=1000
    m,c=gd(data,learning_rate,num_iterations)
    print(m,c)
```
